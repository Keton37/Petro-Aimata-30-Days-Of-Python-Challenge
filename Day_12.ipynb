# Day 12 - List Comprehensions

**Concepts covered:**

- Creating lists using comprehensions
- Conditional comprehensions

---




## Examples

```python
squares = [x*x for x in range(1,21)]
evens = [x for x in range(1,21) if x%2==0]
print(squares[:5], evens[:5])
```
