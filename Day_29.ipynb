```python
# Skeleton for todo app
import json, os
FNAME = "tasks.json"
def load_tasks():
    if os.path.exists(FNAME):
        return json.load(open(FNAME))
    return []
def save_tasks(tasks):
    json.dump(tasks, open(FNAME,"w"), indent=2)
```
