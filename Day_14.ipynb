---



## Example



```python
try:
    x = int("abc")
except ValueError as e:
    print("Conversion error:", e)
finally:
    print("Cleanup if needed")
```

## Exercises

**Exercise 1:** Handle division by zero in a function.

## Solutions

```python
def safe_div(a,b):
    try:
        return a/b
    except ZeroDivisionError:
        return None
print(safe_div(5,0))
```
