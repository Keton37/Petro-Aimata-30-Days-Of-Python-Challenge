# Day 08 - Conditionals

**Concepts covered:**

- if, elif, else
- Comparison operators
- Logical operators and nested conditionals

---

## Examples

```python
n = 7
if n % 2 == 0:
    print("Even")
elif n % 3 == 0:
    print("Divisible by 3")
else:
    print("Odd")
```
