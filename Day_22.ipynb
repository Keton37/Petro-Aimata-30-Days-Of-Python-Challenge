## Exercises

**Exercise 1:** Scrape all headings (h1-h3) from a webpage.

## Solutions

```python
%pip install bs4

import requests
from bs4 import BeautifulSoup
url="https://example.com"
r=requests.get(url)
soup=BeautifulSoup(r.text,"html.parser")
for h in soup.find_all(["h1","h2","h3"]):
    print(h.get_text(strip=True))
```
