## Solutions

```python
# Simple Linear Regression on a small dataset

import numpy as np
from sklearn.linear_model import LinearRegression

# Small sample dataset
# Example: Years of experience vs salary
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)   # Feature (years)
y = np.array([30, 35, 40, 45, 50])             # Target (salary in thousands)

# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Display model coefficients
print("Slope (Coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)

# Make a prediction
experience = 6
predicted_salary = model.predict([[experience]])
print(f"Predicted salary for {experience} years of experience:", predicted_salary[0])

```
