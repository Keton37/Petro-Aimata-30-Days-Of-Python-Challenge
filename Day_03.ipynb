# Day 03 - Strings

**Concepts covered:**

- String operations
- Slicing
- f-strings
- escape characters
- string methods (upper, lower, split)

---



## Examples

```python
text = "Hello, Petro!"
print(text.upper())
print(text[0:5])  # slicing
first_name = "Petro"
last_name = "Aimata"
full = f"{first_name} {last_name}"
print(full)
# escape
print('He said, "Hello!"')
```

## Exercises

**Exercise 1:** Concatenate first and last name.
**Exercise 2:** Slice the first 3 letters of your name.
**Exercise 3:** Split a sentence into words.

## Solutions for the above exercises

```python
first = "Petro"
last = "Longor"
print(first + " " + last)
print(first[:3])
print("I love Python".split())
```
