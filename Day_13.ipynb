# Day 13 - Higher-Order Functions

**Concepts covered:**

- map, filter, reduce
- Passing functions as arguments
- lambda functions

---



## Example



```python
names = ["petro","anna","mike"]
upper = list(map(str.title, names))
print(upper)
evens = list(filter(lambda x: x%2==0, range(10)))
print(evens)
from functools import reduce
total = reduce(lambda a,b: a+b, [1,2,3,4])
print(total)
```
