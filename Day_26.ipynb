## Examples

```python
import matplotlib.pyplot as plt
x = list(range(1,11))
y = [v*v for v in x]
plt.plot(x,y)
plt.title("x vs x^2")
plt.xlabel("x")
plt.ylabel("x^2")
plt.show()
```

## Exercises

**Exercise 1:** Plot a line graph of 1â€“10 vs their squares.

Solutions

```python
import matplotlib.pyplot as plt

# Data
x = list(range(1, 11))          # 1 to 10
y = [i**2 for i in x]           # squares

# Plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o')

# Labels and title
plt.title("Numbers 1â€“10 vs Their Squares")
plt.xlabel("Numbers")
plt.ylabel("Squares")
plt.grid(True)

plt.show()
```
