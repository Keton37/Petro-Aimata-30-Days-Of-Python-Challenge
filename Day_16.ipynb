# Day 16 - Inheritance

**Concepts covered:**

- Parent and child classes
- super()
- Method overriding

---


## Examples

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):
        return f"Hi, I'm {self.name} and I'm {self.age} years old"

class Student(Person):
    def __init__(self, name, age, course):
        super().__init__(name, age)
        self.course = course

    def greet(self):
        return super().greet() + f" and I study {self.course}"

s = Student("Asha", 21, "Data Science")
print(s.greet())
```
