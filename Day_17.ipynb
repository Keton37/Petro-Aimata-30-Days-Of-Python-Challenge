# Day 17 - File Handling

**Concepts covered:**

- open(), read(), write(), with context manager
- Working with text files

---


## Examples

```python
with open("hello.txt","w") as f:
    f.write("Hello, 30 Days of Python!\n")
with open("hello.txt","r") as f:
    print(f.read())
```

## Exercises

**Exercise 1:** Write and read a text file named notes.txt.

## Solutions

```python
with open("notes.txt","w") as f:
    f.write("My notes")
with open("notes.txt") as f:
    print(f.read())
```
